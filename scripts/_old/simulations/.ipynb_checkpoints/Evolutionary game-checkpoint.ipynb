```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.integrate import odeint

plt.style.use('default')

output_dir = "/Users/gelkouh/Library/CloudStorage/OneDrive-Personal/Documents/School/UChicago/Thesis/MINING/coal-mining/output/figures/"
```

```python
# Define functions for governing equations
def xdot_1a(x_1a, x_2a, x_3a, f_1a, c_1a):
    return x_1a * (1 - x_1a) * (((1 - x_2a) * x_3a) + ((1 + f_1a) * x_2a) - c_1a)

def xdot_2a(x_1a, x_2a, x_3a, f_2, f_u, c_2a):
    return x_2a * (1 - x_2a) * ((x_3a * f_2) + ((1 + f_u) * x_1a) - c_2a)

def xdot_3a(x_1a, x_2a, x_3a, f_3, c_3):
    return x_3a * (1 - x_3a) * (((1 - x_2a) * x_1a * f_3 )+ (x_1a * x_2a) - c_3)

def xdot_1b(x_1b, x_2b, x_3b, f_1b, c_1b):
    return xdot_1a(x_1b, x_2b, x_3b, f_1b, c_1b)

def xdot_2b(x_1b, x_2b, x_3b, f_2, c_2b):
    return xdot_2a(x_1b, x_2b, x_3b, f_2, 0, c_2b)

def xdot_3b(x_1b, x_2b, x_3b, f_3, c_3):
    return xdot_3a(x_1b, x_2b, x_3b, f_3, c_3)
```

```python
# Phase portraits of x_1-x_2 space for holding x_3 fixed
def PhasePortrait2D(X , Y, x_dot, y_dot, 
                    title = "", xlab = '$x_1$', ylab = '$x_2$', 
                    density_value = 1, 
                    vline = False, vline_x = 1, 
                    circles = False, C1 = 0, C2 = 0,
                    figsize = (7,7),
                    file_name = False):
    plt.figure(figsize=figsize)
    if vline:
        plt.axvline(x = vline_x, color = "red")
    if circles:
        plt.contour( X , Y , C1, [0] )
        plt.contour( X , Y , C2, [0] )
    plt.streamplot(X, Y, x_dot, y_dot, 
                   density = density_value, 
                   linewidth = 1,
                   color = 'teal')
    plt.title(f'{title}')
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    if file_name != False:
        plt.savefig(f'{file_name}.pdf')
    plt.show()
    
x_1 = np.linspace( 0, 1, 1000 )
x_2 = np.linspace( 0, 1, 1000 )
X_1,X_2 = np.meshgrid( x_1 , x_2 )
```

```python
def TimeSeries3D(state, t, f_1, f_2, f_3, f_u, c_1, c_2, c_3, union = True):
    x_1, x_2, x_3 = state
    if union:
        x_1_output = xdot_1a(x_1a = x_1, x_2a = x_2, x_3a = x_3, 
                             f_1a = f_1, c_1a = c_1)
        x_2_output = xdot_2a(x_1a = x_1, x_2a = x_2, x_3a = x_3, 
                             f_2 = f_2, f_u = f_u, c_2a = c_2)
        x_3_output = xdot_3a(x_1a = x_1, x_2a = x_2, x_3a = x_3, 
                             f_3 = f_3, c_3 = c_3)
    if not union:
        x_1_output = xdot_1b(x_1b = x_1, x_2b = x_2, x_3b = x_3, 
                             f_1b = f_1, c_1b = c_1)
        x_2_output = xdot_2b(x_1b = x_1, x_2b = x_2, x_3b = x_3, 
                             f_2 = f_2, c_2b = c_2)
        x_3_output = xdot_3b(x_1b = x_1, x_2b = x_2, x_3b = x_3, 
                             f_3 = f_3, c_3 = c_3)
    return x_1_output, x_2_output, x_3_output

def TimeSeriesPlot(init_cond, t_vals, args_list, title, y_lim = False, file_name = False):
    solution_timeseries3D = odeint(TimeSeries3D, init_cond, t_vals,
                                   args = args_list)
    plt.figure(figsize = (9,6))
    plt.plot(t_vals, solution_timeseries3D[:,0], label = 'x_1')
    plt.plot(t_vals, solution_timeseries3D[:,1], label = 'x_2')
    plt.plot(t_vals, solution_timeseries3D[:,2], label = 'x_3')
    plt.legend()
    plt.xlabel("t")
    plt.title(f'{title}')
    if y_lim != False:
        plt.ylim(y_lim)
    if file_name != False:
        plt.savefig(f'{file_name}.pdf')
    plt.show()
```
